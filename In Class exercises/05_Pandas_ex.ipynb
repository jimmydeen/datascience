### Exercise 4

Using the weather data, use `df.plot.hist()` to plot a histogram of dewpoint temperature and air temperature. Include both distrubtions on the same plot. See https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.hist.html.

```python
apw = pd.read_csv("IDV60901.94866.csv")
apw.drop(columns=['apparent_t','press_msl','press_tend[80]','rel_hum','vis_km[80]','weather[80]','wind_dir[80]'], inplace=True)
apw.columns = ["DateTime","FullDateTime","Cloud","CloudBase","CloudType","WindGust","AirTemperature","DewPtTemperature","Rain","WindSpeed"]

n_apw= apw.drop(columns= ["DateTime", "FullDateTime", "Cloud", "CloudBase", "CloudType", "WindGust", "Rain", "WindSpeed"])
n_apw.plot.hist(alpha= 0.3);
plt.xlabel("Temperature (Celsius)")


```

### Exercise 5

Using the weather data, calculate the maximum and minimum temperatures for the 25 March. You will need to subset the data and use `df.min()` and `df.max()`.

```python
march_25 = apw[(apw.FullDateTime >= 2021032500000) & (apw.FullDateTime <20210326000000)]

march_25_temp= march_25.drop(columns=["DateTime", "FullDateTime", "Cloud", "CloudBase", "CloudType", "WindGust", "Rain", "WindSpeed"])#Get rid of all columns except the temperature data

print(f"The minimum temperatures for March 25th were Air Temperature: {march_25_temp.min()[0]} Degrees Celsius and Dew Point Temperature: {march_25_temp.min()[1]} Degrees Celsius")
print(f"The maximum temperatures for March 25th were Air Temperature: {march_25_temp.max()[0]} Degrees Celsius and Dew Point Temperature: {march_25_temp.max()[1]} Degrees Celsius")
```
