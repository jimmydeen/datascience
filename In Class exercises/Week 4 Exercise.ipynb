### Exercise 2

The Iris dataset (see pre-studio notebook) has four features corresponding to sepal length, sepal width, petal length and petal width, all measured in cm. Create an overlayed histogram which compares the relative frequency of these four features. Your plot should include an appropriate legend, title and x and y axis labels.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.hist(features[0], density=True, alpha=0.2, histtype='stepfilled', bins=15)
plt.hist(features[1], density=True, alpha=0.2, histtype='stepfilled', bins=15)
plt.hist(features[2], density=True, alpha=0.2, histtype='stepfilled', bins=15)
plt.hist(features[3], density=True, alpha=0.2, histtype='stepfilled', bins=15);
plt.legend(iris.feature_names)
plt.xlabel("Feature Length/Width (cm)")
plt.ylabel("Relative Frequency")
plt.title("Comparison of Iris Features")

```

### Exercise 3

The list `cities` in the following cell contains the population and latitude and longitude of thirteen major cities of Australia. Use this to plot the position of the cities with size proportional to the population (used `scaled_pop`), and coloured dependent on whether they are the national capital, state capital or regional cities. The legend should explain this last feature.  

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

cities = [['Sydney','Melbourne','Brisbane','Perth','Adelaide', 'Gold Coast',
           'Canberra','Newcastle','Wollongong','Logan City','Geelong','Hobart', 'Darwin'],
          [4627345, 4246375, 2189878, 1896548, 1225235, 591473, 367752, 
           308308, 292190, 282673, 226034, 216656, 147255],
          [-33.868, -37.814, -27.468, -31.952, -34.929, -28, -35.283, 
           -32.93, -34.424, -27.639, -38.147, -42.879, -12.462827],
          [151.207, 144.963, 153.028, 115.861, 138.599, 153.431, 149.128,
           151.78, 150.893, 153.109, 144.361, 147.329, 130.841782],
          ['b','b','b','b','b','g','r','g','g','g','g','b','b']]

scale_factor = 0.0001

p = cities[1]

# scaled_pop = [scale_factor*x for x in cities[1]]   - Used in for loop below

n_cap = "Canberra" 
s_cap = ["Sydney", "Melbourne", "Brisbane", "Perth", "Adelaide", "Hobart", "Darwin"]
reg = ["Gold Coast", "Newcastle", "Wollongong", "Logan City", "Geelong"]

n_cap_col = "red"
s_cap_col = "gold"
reg_col = "blue"


color_coord = []
ydata= cities[2]
xdata=cities[3]
a = 0




for city in cities[0]:
    
    scaled_pop= p[a]*scale_factor

    if city in n_cap:
        color_coord = n_cap_col
    elif city in s_cap:
        color_coord = s_cap_col
    elif city in reg:
        color_coord = reg_col
    plt.scatter(xdata[a],ydata[a], s=scaled_pop, alpha = 0.4, color = color_coord)
    
    a += 1
plt.title("Position of Australian Cities with Proportional Size to Population")
```

### Exercise 4

The population of the five most populous Australian cities at the last three census is recorded in the following cell. Plot a side-by-side bar chart to show the change in the populations. The chart should be grouped by each city and show each census in different colours. The legend should explain the colours for the census. Your plot should include an appropriate title.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

cities = ['Sydney','Melbourne','Brisbane','Perth','Adelaide']
Pop_2016 = [4321535,4196198,2054614,1874578,1165632]
Pop_2011 = [3908642,3707530,1874427,1627576,1103979]
Pop_2006 = [3641422,3371888,1676389,1256035,1040719]
data = np.array([Pop_2016, Pop_2011, Pop_2006])


X = np.arange(len(cities))
bar_width = .25

plt.bar(X + 0.00, data[0,:], color = 'orange', width = bar_width, label='2016')
plt.bar(X + 0.25, data[1,:], color = 'green', width = bar_width, label='2011')
plt.bar(X + 0.50, data[2,:], color = 'gold', width = bar_width, label='2006')
plt.xticks(X + .25, cities)
plt.ylabel("Population (millions)")
plt.legend()

plt.title('Change in Population of the 5 Largest Australian Cities (Past 3 Census)');
```
