```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

rainfall = pd.read_csv('MelboureMonthlyRainfall.csv')
maxtemp = pd.read_csv('MelbourneMonthlyMaxT.csv')
mintemp = pd.read_csv('MelbourneMonthlyMinT.csv')
dmi = pd.read_csv('dmi.had.long.data.csv')
sam = pd.read_csv('sam.csv')
soi = pd.read_csv('soi.csv')

###DATA CLEANING###

rainfall = rainfall.drop(['Product code','Station Number'],1)
maxtemp = maxtemp.drop(['Product code','Station Number'],1)
mintemp = mintemp.drop(['Product code','Station Number'],1)

soi = soi.dropna()
sam = sam.dropna()
dmi = dmi.dropna()
rainfall = rainfall.dropna()
maxtemp = maxtemp.dropna()
mintemp = mintemp.dropna()

sam = sam.rename(columns={'Unnamed: 0': 'YEAR'})
dmi = dmi.rename(columns={'Unnamed: 0': 'YEAR'})


maxtemp = maxtemp[maxtemp['Year']> 1956].reset_index().drop('index', axis =1).rename(columns={'Annual': 'AnnualMaxTemp'})

rainfall= rainfall[rainfall['Year']> 1956].reset_index().drop('index', axis =1).rename(columns={'Annual': 'AnnualRainfall'})


mintemp = mintemp[mintemp['Year']> 1956].reset_index().drop('index', axis =1).rename(columns={'Annual': 'AnnualMinTemp'})


dmi = dmi[dmi['YEAR']> 1956].reset_index().drop('index', axis =1).drop(index=65)



soi = soi[soi['YEAR']> 1956].reset_index().drop('index', axis =1)

sam_int_year = sam['YEAR'].astype(int)

sam = pd.merge(sam_int_year, sam, how = 'left')

rainfall.set_index('Year', inplace = True)
mintemp.set_index('Year', inplace = True)
maxtemp.set_index('Year', inplace = True)

dmi.set_index('YEAR', inplace = True)
sam.set_index('YEAR', inplace = True)
soi.set_index('YEAR', inplace = True)

rainfall.columns = [x.upper() for x in rainfall.columns]
mintemp.columns = [x.upper() for x in mintemp.columns]
maxtemp.columns = [x.upper() for x in maxtemp.columns]

###END DATA CLEANING###
```

```python
#Add Annual Means Column
soi['AnnualSOI']= soi.mean(numeric_only=True, axis=1)
dmi['AnnualDMI']= dmi.mean(numeric_only=True, axis=1)
sam['AnnualSAM']= sam.mean(numeric_only=True, axis=1)
```
